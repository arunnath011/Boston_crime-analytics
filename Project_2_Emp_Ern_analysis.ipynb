```python
import pandas as pd
import numpy as np 
#1
#df=pd.read_csv('e2e2c23a-6fc7-4456-8751-5321d8aa869b.csv',index_col='NAME')
df=pd.read_csv('employee-earnings-combined2015-2020.csv')
```

```python
df.head()
```

```python
df.fillna(np.nan)
```

```python
df=df.drop(columns=['Unnamed: 0'])
```

```python
df['REGULAR']=df['REGULAR'].replace('[^0-9\.-]', '',regex=True)
```

```python
df.head()
```

```python
df['RETRO']=df['RETRO'].replace('[^0-9\.-]', '',regex=True)
df['OTHER']=df['OTHER'].replace('[^0-9\.-]', '',regex=True)
df['OVERTIME']=df['OVERTIME'].replace('[^0-9\.-]', '',regex=True)
df['INJURED']=df['INJURED'].replace('[^0-9\.-]', '',regex=True)
df['DETAILS']=df['DETAILS'].replace('[^0-9\.-]', '',regex=True)
df['QUINN/EDUCATION INCENTIVE']=df['QUINN/EDUCATION INCENTIVE'].replace('[^0-9\.-]', '',regex=True)
df['TOTAL EARNINGS']=df['TOTAL EARNINGS'].replace('[^0-9\.-]', '',regex=True)
```

```python
df.head()
```

```python
df[['REGULAR','RETRO','OTHER','OVERTIME','INJURED','DETAILS','QUINN/EDUCATION INCENTIVE','TOTAL EARNINGS']]=df[['REGULAR','RETRO','OTHER','OVERTIME','INJURED','DETAILS','QUINN/EDUCATION INCENTIVE','TOTAL EARNINGS']].replace('',np.nan )
```

```python
df[['REGULAR','RETRO','OTHER','OVERTIME','INJURED','DETAILS','QUINN/EDUCATION INCENTIVE','TOTAL EARNINGS']]=df[['REGULAR','RETRO','OTHER','OVERTIME','INJURED','DETAILS','QUINN/EDUCATION INCENTIVE','TOTAL EARNINGS']].replace('-',np.nan )
```

```python
df.isnull().sum(axis=0)
```

```python
df[['REGULAR','RETRO','OTHER','OVERTIME','INJURED','DETAILS','QUINN/EDUCATION INCENTIVE','TOTAL EARNINGS']]=df[['REGULAR','RETRO','OTHER','OVERTIME','INJURED','DETAILS','QUINN/EDUCATION INCENTIVE','TOTAL EARNINGS']].fillna(0)
```

```python
df.isnull().sum(axis=0)
```

```python
df[['REGULAR','RETRO','OTHER','OVERTIME','INJURED','DETAILS','QUINN/EDUCATION INCENTIVE','TOTAL EARNINGS']]=df[['REGULAR','RETRO','OTHER','OVERTIME','INJURED','DETAILS','QUINN/EDUCATION INCENTIVE','TOTAL EARNINGS']].astype(float)
```
